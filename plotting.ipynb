```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def plot_file(filename, pt=False):
    vals = np.loadtxt(filename)
    
    if pt: print(vals)
    for i in range(1, len(vals)):
        plt.plot(vals[0], vals[i])
#     return vals
```

```python
plot_file("values.txt")
# print(vals)
plt.savefig("small.png")
```
